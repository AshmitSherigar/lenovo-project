```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("power_emergency_raw_dataset 1.csv")
df.head(20)
```

```python
df.info()
```

```python
df_s1 = df[df['Server_ID'] == 'S1'].copy()
df_s2 = df[df['Server_ID'] == 'S2'].copy()
df_s3 = df[df['Server_ID'] == 'S3'].copy()
```

```python
df_s1.shape
```

```python
df_s1['Timestamp'] = pd.to_datetime(df_s1['Timestamp'])
df_s1 = df_s1.sort_values(by='Timestamp')
```

```python
df_s1.head(20)
```

```python
df_s1.info()
```

```python
df_s1['hour'] = df_s1['Timestamp'].dt.hour
df_s1['minute'] = df_s1['Timestamp'].dt.minute
df_s1['day_of_week'] = df_s1['Timestamp'].dt.dayofweek
```

```python
df_s1['day_of_week']
```

```python
df_s1['cpu_change'] = df_s1['CPU_Utilization_%'].diff()
df_s1['memory_change'] = df_s1['Memory_Utilization_%'].diff()
df_s1['power_change'] = df_s1['Power_Usage_Watts'].diff()
```

```python
df_s1
```

```python
df_s1['cpu_rolling_mean'] = df_s1['CPU_Utilization_%'].rolling(5).mean()
df_s1['power_rolling_mean'] = df_s1['Power_Usage_Watts'].rolling(5).mean()
df_s1['power_deviation'] = df_s1['Power_Usage_Watts'] - df_s1['power_rolling_mean']
df_s1['cpu_deviation'] = df_s1['CPU_Utilization_%'] - df_s1['cpu_rolling_mean']
```

```python
df_s1
```

```python
df_s1 = df_s1.dropna()
```

```python
features = [
    'CPU_Utilization_%',
    'Memory_Utilization_%',
    # 'Power_Usage_Watts',

    'cpu_change',
    'memory_change',
    'power_change',

    'cpu_rolling_mean',
    'power_rolling_mean',

    'cpu_deviation',
    'power_deviation',

    'hour',
    'day_of_week'
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_s1[features])
```

```python
# train_size = int(len(df_s1) * 0.7)
#
# X_train = X_scaled[:train_size]
# X_test = X_scaled[train_size:]
# from sklearn.ensemble import IsolationForest
#
# model = IsolationForest(
#     n_estimators=150,
#     contamination=0.01,
#     random_state=42
# )
#
# model.fit(X_train)
#
# df_s1['anomaly'] = model.decision_function(X_scaled)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(
    n_estimators=150,
    contamination=0.05,
    random_state=42
)

df_s1['anomaly'] = model.fit_predict(X_scaled)
```

```python
df_s1
```

```python
df_s1['anomaly_flag'] = df_s1['anomaly'].apply(lambda x: 1 if x == -1 else 0)
```

```python
df_s1.shape
```

```python
print("df_s1 shape:", df_s1.shape)
print("Power length:", len(df_s1['Power_Usage_Watts']))
```

```python
df_s1 = df_s1.sort_values(by='Timestamp').reset_index(drop=True)
```

```python
scores = model.decision_function(X_scaled)
df_s1['anomaly_score'] = scores

threshold = np.percentile(scores, 5)  # top 4% strongest anomalies
df_s1['strong_anomaly'] = df_s1['anomaly_score'] < threshold
```

```python
start = 0
end = 9000 # change this window to explore

subset = df_s1.iloc[start:end]
```

```python

plt.figure(figsize=(14,5))

plt.plot(subset['Timestamp'], subset['Power_Usage_Watts'], linewidth=1)

plt.scatter(
    subset['Timestamp'][subset['strong_anomaly']],
    subset['Power_Usage_Watts'][subset['strong_anomaly']],
    color='red',
    s=40,
    label='Anomaly'
)

plt.title("Power Usage with Detected Anomalies (Server S1)")
plt.xlabel("Time")
plt.ylabel("Power (Watts)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
df_s1.sample(10)
```

```python
# Compare the mean values of features for Normal vs. Anomalous points
profiling_metrics = df_s1.groupby('strong_anomaly')[['Power_Usage_Watts', 'CPU_Utilization_%', 'cpu_change']].mean()
print(profiling_metrics)
```

```python
anomaly_count = df_s1['strong_anomaly'].sum()
total_count = len(df_s1)
anomaly_rate = (anomaly_count / total_count) * 100

print(f"Total Anomalies: {anomaly_count}")
print(f"Anomaly Rate: {anomaly_rate:.2f}%")
```

```python
# Copy original data
df_test = df_s1.copy()

# Inject synthetic anomalies
np.random.seed(42)
n_anomalies = 50

indices = np.random.choice(len(df_test), n_anomalies, replace=False)

# Create spikes in power
df_test.loc[indices, 'Power_Usage_Watts'] *= 1.8

# Recompute dependent features
df_test['power_change'] = df_test['Power_Usage_Watts'].diff()
df_test['power_rolling_mean'] = df_test['Power_Usage_Watts'].rolling(5).mean()
df_test['power_deviation'] = df_test['Power_Usage_Watts'] - df_test['power_rolling_mean']

df_test = df_test.dropna()

# Scale
X_test_scaled = scaler.transform(df_test[features])

# Score
scores_test = model.decision_function(X_test_scaled)

# Detect anomalies
threshold = np.percentile(scores_test, 4)
df_test['detected'] = scores_test < threshold

# Evaluation
detected_rate = df_test.loc[indices, 'detected'].mean()

print(f"Injected anomalies: {n_anomalies}")
print(f"Detection rate: {detected_rate * 100:.2f}%")
```

```python
results = []

for seed in range(5):
    model_temp = IsolationForest(contamination=0.01, random_state=seed)
    preds = model_temp.fit_predict(X_scaled)
    results.append(preds)

# Compare similarity
for i in range(1, 5):
    similarity = np.mean(results[0] == results[i])
    print(f"Similarity with run {i}: {similarity:.3f}")
```
